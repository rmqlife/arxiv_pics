```python
from util import *
files = get_filelist('./pdf (copy)/',['png'])
print(files)
# files=files[:1]
# print(files)
for fn in files:
    print(fn, fn[:-3]+'jpg')
    print(fn)
    im = cv2.imread(fn, cv2.IMREAD_UNCHANGED)
    show(im)
    print(im.shape)
    while np.max(im.shape)>1000:
        im = cv2.pyrDown(im)
    if len(im.shape)>2 and im.shape[2]==4:
        im_new = np.zeros((im.shape[0], im.shape[1], 3))
        print(im_new.shape)
        for i in range(im.shape[0]):
            for j in range(im.shape[1]):
                im_new[i,j]=im[i,j,3]/255.0*im[i,j,:3].astype('float')+(1-im[i,j,3]/255.0)*np.array([255,255,255])
        im=im_new.astype('uint8')
        show(im)
# #     im = im[:,:,:3]            
    cv2.imwrite(fn[:-3]+'jpg', im)
    im = cv2.imread(fn[:-3]+'jpg')
    cv2.imwrite(fn[:-3]+'png',im)
#     show(im)
# print(im[0,0])
```
